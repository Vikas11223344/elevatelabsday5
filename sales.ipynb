```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load CSV file
df = pd.read_csv('sales.csv')
print("✅ Data Loaded Successfully")
print(df.head())

# Step 3: Basic Info
print(df.info())
print(df.describe())

# Step 4: Total Sales
total_sales = df['Sales'].sum()
print(f"Total Sales: {total_sales}")

# Step 5: Sales by Product
sales_by_product = df.groupby('Product')['Sales'].sum()
print(sales_by_product)

# Step 6: Sales by Region
sales_by_region = df.groupby('Region')['Sales'].sum()
print(sales_by_region)

# Step 7: Visualization - Sales by Product
sales_by_product.plot(kind='bar', color='skyblue', title='Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()

# Step 8: Visualization - Sales by Region
sales_by_region.plot(kind='pie', autopct='%1.1f%%', title='Sales by Region')
plt.ylabel('')
plt.show()

# Step 9: Convert Date to datetime and analyze monthly sales
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')

monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
